## **Random Forest for Classification**

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# Load data
df = pd.read_csv('iris_2D.csv')

# Get data
x_data = df[['Petal_Length', 'Petal_Width']].to_numpy()
x_data = x_data.reshape(6, 2)
y_data = df['Label'].to_numpy(dtype=np.uint8)

# Define model
rf_classifier = RandomForestClassifier(
    n_estimators=3,
    max_features=1,
    max_depth=1,
    criterion='entropy',
    max_samples=6
)

# Train model
rf_classifier.fit(x_data, y_data)

# Predict
x_test = np.array([[2.7, 0.8]])
y_predicted = rf_classifier.predict(x_test)
y_predicted
```

## **Random Forest for Regression**

## **AdaBoost for Classification**

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import AdaBoostClassifier

# Load data
df = pd.read_csv('iris_2D.csv')

# Get data
x_data = df[['Petal_Length', 'Petal_Width']].to_numpy()
x_data = x_data.reshape(6, 2)
y_data = df['Label'].to_numpy(dtype=np.uint8)

# Define model
ab_classifier = AdaBoostClassifier(
    n_estimators=3,
    random_state=0
)

# Train model
ab_classifier.fit(x_data, y_data)

# Predict
x_test = np.array([[2.7, 0.8]])
y_pred = ab_classifier.predict(x_test)
y_pred
```

```python
plt.figure(figsize=(6,6))
_ = tree.plot_tree(
    ab_classifier.estimators_[0],
    feature_names=['Petal_Length', 'Petal_Width'],
    filled=True
)
```

```python
viz_model = dtreeviz.model(
    ab_classifier.estimators_[0],
    X_train=x_data,
    y_train=y_data,
    feature_names=['Petal_Length', 'Petal_Width'],
    target_name='Label'
)

viz_view = viz_model.view(fontname="monospace")
viz_view.save("./iris_1.svg")
viz_view
```

```python
plt.figure(figsize=(6,6))
_ = tree.plot_tree(
    ab_classifier.estimators_[1],
    feature_names=['Petal_Length', 'Petal_Width'],
    filled=True
)
```

```python
viz_model = dtreeviz.model(
    ab_classifier.estimators_[1],
    X_train=x_data,
    y_train=y_data,
    feature_names=['Petal_Length', 'Petal_Width'],
    target_name='Label'
)

viz_view = viz_model.view(fontname="monospace")
viz_view.save("./iris_1.svg")
viz_view
```
